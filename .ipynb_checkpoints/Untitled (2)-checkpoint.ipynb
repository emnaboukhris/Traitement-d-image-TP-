```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
def readpgm(name):
    with open(name) as f:
        lines = f.readlines()
    
    
    entete=[]
    for l in list(lines)[0:4]:
        #print(l)
        entete.append(l)
    
    
    data = []
    for line in lines[2:4]:
        data.extend([int(c) for c in line.split()])
    
   
    return (im.imread(name),(data[1],data[0]),data[2],entete)

data = readpgm('./chat.pgm')
print(data)
plt.imshow(np.reshape(data[0],data[1])) # Usage example
```

```python
import numpy as np
import matplotlib.pyplot as plt
def writePGM(name,img):
    with open(name, 'w') as f:
        f.write('P2\n')
        f.write('#--------\n')
        f.write(str(img.shape[1])+' '+str(img.shape[0])+'\n')
        f.write(str(np.max(img))+'\n')
        for l in img:
            f.write(' '.join(map(str, l)))
            f.write('\n')
            
writePGM('out.pgm',readpgm('./chat.pgm')[0])
data = readpgm('./out.pgm')
print(data)
plt.imshow(np.reshape(data[0],data[1]))  
```

```python
import numpy as np
import matplotlib.pyplot as plt
def histogramme(name):
    image=readpgm(name)[0]
  
    histogramme=np.zeros(256, int)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            histogramme[image[i,j]]=histogramme[image[i,j]]+1
    return histogramme
def histogrammeCummule(name):
    histCum=histogramme(name)
    for i in range(1,len(histCum)):
        histCum[i]+=histCum[i-1]
    return histCum
plt.plot(histogramme('./chat.pgm'))
plt.show()
plt.hist(histogramme('./chat.pgm'))
#print(histogramme('./chat.pgm')) 
np.size(histogramme('./chat.pgm'))
plt.show()
plt.plot(histogrammeCummule('./chat.pgm'))
plt.show()
#print(histogrammeCummule('./chat.pgm'))
```

```python
# This is a sample Python script.
import matplotlib.image as mpimg
import numpy as np
import numpy as np
import random
from matplotlib import image
from numpy import asarray
from math import *


def pmgBruit(name):
    with open(name) as f:
        data = image.imread(name)
        matrice0 = asarray(data)
        matrice =np.copy(matrice0)
        for i in range(matrice.shape[0]) :
            for j in range(matrice.shape[1]):
                randomNumber = random.randint(0, 20)
                if randomNumber == 0:
                    matrice[i][j] = 0
                if randomNumber == 20:
                    matrice[i][j] = 255
    return matrice

def convolution2D(X,H):
    s = X.shape
    py = int((H.shape[0]-1)/2)
    px = int((H.shape[1]-1)/2)
    Y = X.copy()
    xmax= int(s[1]-px) ;
    ymax= int(s[0]-py)
    for i in range(px,xmax):
        for j in range(py,ymax):
            somme = 0.0
            for k in range(-px,px+1):
                for l in range(-py,py+1):
                    somme += X[j+l][i+k]*H[l+py][k+px]
            Y[j][i] = somme
    return Y

def filtreMedian(X,N):
    s = X.shape
    py = int((N-1)/2)
    px = int((N-1)/2)
    Y = X.copy()
    iMed = int(N / 2)

    xmax= int(s[1]-px) ;
    ymax= int(s[0]-py)
    for i in range(px,xmax):
        for j in range(py,ymax):
            med = []
            for k in range(-px,px+1):
                for l in range(-py,py+1):
                    med.append(Y[k][l])
            med.sort()
            Y[j][i] = med[iMed]
    return Y

def produitSNR(X,H):
    py = int(X.shape[1])
    px = int(X.shape[0])
    mean = X.mean() ;
    numerateur =0
    denominateur =int(0)

    for i in range(0,px):
        for j in range(0,py):
            numerateur = numerateur +(X[i][j]-mean)*(X[i][j]-mean)
            denominateur =int( denominateur+((H[i][j]-X[i][j])*(H[i][j]-X[i][j])))
    snr =sqrt(numerateur/denominateur)
    return snr



data = pmgBruit('./chat.pgm')
filterMoy= np.ones((3, 3)) * 1.0 / 9
filtredDataMoy=convolution2D(data,filterMoy)
filtredDataMed=filtreMedian(data,3)
filtrePasseHaut= np.reshape([0 ,1/9 ,0,-1/9 ,5/9,-1/9, 0 ,-1/9,0], (3, 3))
filtredDataPassHaut=convolution2D(data,filtrePasseHaut)
plt.imshow(data)
plt.show()
plt.imshow(filtredDataPassHaut)
plt.show()
print(filtredDataPassHaut)
print("this is filtre Passe Haut")
print(filtredDataPassHaut)
print("this is mean")
print(data.mean())
snr1=produitSNR(data,filtredDataMoy)
snr2=produitSNR(data,filtredDataMed)
print("the snr")
print(snr1)
print(snr2)

   
```

```python
from tkinter import *
window=Tk()
#personaliser window
window.title("traitement d'image")
window.geometry("1080x720")
window.minsize(480,360)
window.iconbitmap("hoko-logo-500.ico")
window.config(background='#2c2c2c')

#create frame
frame=Frame(window,bg='#2c2c2c')
def welcomeWindow():
    def nextWindow():
        frame.destroy()
        def equalizationPage():
            equalisationFrame=Frame(mainFrame,bg='#2c2c2c')
            label_title_eq=Label(equalisationFrame,text="Image Equalization",font=("Garamond",40),bg='#2c2c2c',fg='#e9e9e9')
            label_title_eq.pack()
            
            equalisationFrame.pack(pady=20)
        def contrastPage():
            contrastFrame=Frame(mainFrame,bg='#2c2c2c')
            lb=Label(contrastFrame,text='contrast')
            lb.pack()
            contrastFrame.pack(pady=20)
        def filterPage():
            filterFrame=Frame(mainFrame,bg='#2c2c2c')
            lb=Label(filterFrame,text='filter')
            lb.pack()
            filterFrame.pack(pady=20)
        def deletePages():
            for frame in mainFrame.winfo_children():
                frame.destroy()
            
        def hide_indicators():
            equal_indicate.config(bg='#f3efe0')
            contrast_indicate.config(bg='#f3efe0')
            filter_indicate.config(bg='#f3efe0')
        def indicate(lb,page):
            hide_indicators()
            lb.config(bg='#2c2c2c')
            deletePages()
            page()
            
        optionFrame=Frame(window,bg='#f3efe0')
        optionFrame.pack(side=LEFT)
        optionFrame.pack_propagate(False)
        optionFrame.configure(width=200,height=720)
        equal_btn=Button(optionFrame,text='Image Equalization',font=("Garamond",15),bg='#f3efe0',fg='#2c2c2c',bd=0,
                        command=lambda:indicate(equal_indicate,equalizationPage))
        equal_btn.place(x=10,y=30,width=180)
        equal_indicate=Label(optionFrame,text='',bg='#f3efe0')
        equal_indicate.place(x=3,y=30,width=5,height=35)
        contrast_btn=Button(optionFrame,text='Image contraction',font=("Garamond",15),bg='#f3efe0',fg='#2c2c2c',bd=0,
                           command=lambda:indicate(contrast_indicate,contrastPage))
        contrast_btn.place(x=10,y=90,width=180)
        contrast_indicate=Label(optionFrame,text='',bg='#f3efe0')
        contrast_indicate.place(x=3,y=90,width=5,height=35)
        filter_btn=Button(optionFrame,text='Image filtration',font=("Garamond",15),bg='#f3efe0',fg='#2c2c2c',bd=0,
                         command=lambda:indicate(filter_indicate,filterPage))
        filter_btn.place(x=10,y=150,width=180)
        filter_indicate=Label(optionFrame,text='',bg='#f3efe0')
        filter_indicate.place(x=3,y=150,width=5,height=35)
        
        mainFrame=Frame(window,bg='#2c2c2c',highlightbackground='pink',highlightthickness=2)
        mainFrame.pack(side=LEFT)
        mainFrame.pack_propagate(False)
        mainFrame.configure(height=720,width=1080)
    #add title
    label_title=Label(frame,text="Imag-in",font=("Garamond",40),bg='#2c2c2c',fg='#e9e9e9')
    label_title.pack()
    label_subtitle=Label(frame,text="Here you can process your image perfectly :)",font=("Garamond",20),bg='#2c2c2c',fg='#e9e9e9')
    label_subtitle.pack()
    enter_button=Button(frame,
                        text="Let's Go ->",font=("Garamond",15),bg='#2c2c2c',fg='#e9e9e9',borderwidth = 1,
                        command=nextWindow)
    enter_button.pack(pady=25)
    frame.pack(expand=YES)
welcomeWindow()
window.mainloop()
```
