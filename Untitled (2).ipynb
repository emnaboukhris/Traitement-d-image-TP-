```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
def readpgm(name):
    with open(name) as f:
        lines = f.readlines()
    
    
    entete=[]
    for l in list(lines)[0:4]:
        #print(l)
        entete.append(l)
    
    
    data = []
    for line in lines[2:4]:
        data.extend([int(c) for c in line.split()])
    
   
    return (im.imread(name),(data[1],data[0]),data[2],entete)

data = readpgm('./chat.pgm')
print(data)
plt.imshow(np.reshape(data[0],data[1])) # Usage example
```

```python
import numpy as np
import matplotlib.pyplot as plt
def writePGM(name,img):
    with open(name, 'w') as f:
        f.write('P2\n')
        f.write('#--------\n')
        f.write(str(img.shape[1])+' '+str(img.shape[0])+'\n')
        f.write(str(np.max(img))+'\n')
        for l in img:
            f.write(' '.join(map(str, l)))
            f.write('\n')
            
writePGM('out.pgm',readpgm('./chat.pgm')[0])
data = readpgm('./out.pgm')
print(data)
plt.imshow(np.reshape(data[0],data[1]))  
```

```python
import numpy as np
import matplotlib.pyplot as plt
def histogramme(name):
    image=readpgm(name)[0]
  
    histogramme=np.zeros(256, int)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            histogramme[image[i,j]]=histogramme[image[i,j]]+1
    return histogramme
def histogrammeCummule(name):
    histCum=histogramme(name)
    for i in range(1,len(histCum)):
        histCum[i]+=histCum[i-1]
    return histCum
plt.plot(histogramme('./chat.pgm'))
plt.show()
plt.hist(histogramme('./chat.pgm'))
#print(histogramme('./chat.pgm')) 
np.size(histogramme('./chat.pgm'))
plt.show()
plt.plot(histogrammeCummule('./chat.pgm'))
plt.show()
#print(histogrammeCummule('./chat.pgm'))
```

```python
# This is a sample Python script.
import matplotlib.image as mpimg
import numpy as np
import numpy as np
import random
from matplotlib import image
from numpy import asarray
from math import *


def pmgBruit(name):
    with open(name) as f:
        data = image.imread(name)
        matrice0 = asarray(data)
        matrice =np.copy(matrice0)
        for i in range(matrice.shape[0]) :
            for j in range(matrice.shape[1]):
                randomNumber = random.randint(0, 20)
                if randomNumber == 0:
                    matrice[i][j] = 0
                if randomNumber == 20:
                    matrice[i][j] = 255
    return matrice

def convolution2D(X,H):
    s = X.shape
    py = int((H.shape[0]-1)/2)
    px = int((H.shape[1]-1)/2)
    Y = X.copy()
    xmax= int(s[1]-px) ;
    ymax= int(s[0]-py)
    for i in range(px,xmax):
        for j in range(py,ymax):
            somme = 0.0
            for k in range(-px,px+1):
                for l in range(-py,py+1):
                    somme += X[j+l][i+k]*H[l+py][k+px]
            Y[j][i] = somme
    return Y

def filtreMedian(X,N):
    s = X.shape
    py = int((N-1)/2)
    px = int((N-1)/2)
    Y = X.copy()
    iMed = int(N / 2)

    xmax= int(s[1]-px) ;
    ymax= int(s[0]-py)
    for i in range(px,xmax):
        for j in range(py,ymax):
            med = []
            for k in range(-px,px+1):
                for l in range(-py,py+1):
                    med.append(Y[k][l])
            med.sort()
            Y[j][i] = med[iMed]
    return Y

def produitSNR(X,H):
    py = int(X.shape[1])
    px = int(X.shape[0])
    mean = X.mean() ;
    numerateur =0
    denominateur =int(0)

    for i in range(0,px):
        for j in range(0,py):
            numerateur = numerateur +(X[i][j]-mean)*(X[i][j]-mean)
            denominateur =int( denominateur+((H[i][j]-X[i][j])*(H[i][j]-X[i][j])))
    snr =sqrt(numerateur/denominateur)
    return snr



data = pmgBruit('./chat.pgm')
filterMoy= np.ones((3, 3)) * 1.0 / 9
filtredDataMoy=convolution2D(data,filterMoy)
filtredDataMed=filtreMedian(data,3)
filtrePasseHaut= np.reshape([0 ,1/9 ,0,-1/9 ,5/9,-1/9, 0 ,-1/9,0], (3, 3))
filtredDataPassHaut=convolution2D(data,filtrePasseHaut)
plt.imshow(data)
plt.show()
plt.imshow(filtredDataPassHaut)
plt.show()
print(filtredDataPassHaut)
print("this is filtre Passe Haut")
print(filtredDataPassHaut)
print("this is mean")
print(data.mean())
snr1=produitSNR(data,filtredDataMoy)
snr2=produitSNR(data,filtredDataMed)
print("the snr")
print(snr1)
print(snr2)

   
```
